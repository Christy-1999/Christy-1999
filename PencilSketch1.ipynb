```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('oaaaa.jpg')
```

```python
img = cv2.resize(img,None, fx=0.3, fy = 0.3)
cv2_imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
inv_img = cv2.bitwise_not(gray)
cv2_imshow(inv_img)
```

```python
blurred_inv_img = cv2.GaussianBlur(inv_img, (21, 21), 0)
cv2_imshow(blurred_inv_img)
```

```python
blurred_gray_img = cv2.bitwise_not(blurred_inv_img)
cv2_imshow(blurred_gray_img)
```

```python
pencil_sketch = cv2.divide(gray, blurred_gray_img, scale=256.0)
cv2_imshow(pencil_sketch)
```
